```python
from file import load
```

```python
data = load('USD')
details = data['ticker_details']
print(details.ticker_root, details.name)
for w in data['options_weekly']:
  print(w['day'].date, w['day'])
  for p in w['options']:
    c = p['contract']
    print('  ', c.expiration_date, c.strike_price)
    for h in p['history']:
      print('    ', h.date, h)
```
