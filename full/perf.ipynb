```python
from statistics import mean
from datetime import date
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt

from file import load
from model import create_transactions
```

```python
plt.figure(figsize=(16,9))

for symbol in ['SMH','USD','TQQQ','QQQ']:
    data = load(symbol)
    tx = create_transactions(data)
    weeks = set(t['weeks'] for t in tx if t['weeks'] < 40)
    average_ratios = {w: mean(t['profit_ratio'] for t in tx if t['weeks'] == w) for w in weeks}
    plt.plot(average_ratios.keys(), average_ratios.values(), marker='.', label=symbol)

plt.xlabel('Weeks')
plt.ylabel('Average Ratio')
plt.title('Average Ratios')
plt.grid()
plt.legend()
plt.show()
```
