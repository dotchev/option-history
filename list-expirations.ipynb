```python
import os
import pickle
import sys
import pandas as pd
from datetime import date
from itertools import pairwise
from statistics import quantiles, mean, median
from time import sleep

from polygon import RESTClient
from polygon.rest.models import Agg

from model import OptionData, WeekData, History, save_history
from history import list_stock_history

client = RESTClient()  # POLYGON_API_KEY environment variable is used

symbol = 'USD'
```

```python
to_date = date.today()
from_date = to_date.replace(year=to_date.year - 2)

stock_history = list_stock_history(symbol, from_date, to_date)
```

```python
for start_day in stock_history:
    call_contracts = list(client.list_options_contracts(
        symbol,
        contract_type='call',
        as_of=start_day.date,
        strike_price_gte=start_day.close,
        strike_price_lte=start_day.close * 1.1,
        sort='expiration_date',
        order='asc',
        limit=1000))
    expiration_dates = []
    for c in call_contracts:
        if not c.expiration_date in expiration_dates:
            expiration_dates.append(c.expiration_date)
    exp_weeks = [round((date.fromisoformat(d) - start_day.date).days / 7) for d in expiration_dates]
    print(start_day.date, *[f'{w:3d}' for w in exp_weeks])
```
